```python
import numpy as np

f = open("data_file/line_fit_sample_data_with_delimiter_comma.txt", "r")

datas = []
x_data=[]
y_data=[]

lines = f.readlines()
for line in lines:
    d=line[:-1].split(",")
    x_data.append(float(d[0]))
    y_data.append(float(d[1]))

x_data=np.array(x_data)
y_data=np.array(y_data)

f.close()
```

```python
import matplotlib.pyplot as plt

slope=0.16629283676471487
intercept=0.2254324371848604

plt.title("ransac")
plt.scatter(x_data, y_data)
plt.plot(x_data, slope * x_data + intercept, color='red')
plt.show()
```
