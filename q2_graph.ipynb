```python
import numpy as np

f = open("data_file/plane_fit_sample_data_with_delimiter_comma.txt", "r")

datas = []
x_data=[]
y_data=[]
z_data=[]

lines = f.readlines()
for line in lines:
    d=line[:-1].split(",")
    x_data.append(float(d[0]))
    y_data.append(float(d[1]))
    z_data.append(float(d[2]))

x_data=np.array(x_data)
y_data=np.array(y_data)
z_data=np.array(z_data)

f.close()
```

```python
import matplotlib.pyplot as plt

normal_vector=[13.568494793850007,30.36767659612,12.05201604797]
plane_d=-0.5176533300824246

fig1 = plt.figure()
ax1 = fig1.add_subplot(111, projection='3d')

scatter_data = ax1.scatter(x_data, y_data, z_data, label='데이터', alpha=0.7)

xx, yy = np.meshgrid(np.linspace(min(x_data), max(x_data), 10), np.linspace(min(y_data), max(y_data), 10))
zz_ransac = (-normal_vector[0] * xx - normal_vector[1] * yy - plane_d) / normal_vector[2]
ax1.plot_surface(xx, yy, zz_ransac, color='green', alpha=0.5)

plt.show()
```
