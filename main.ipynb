```python
import matplotlib.pyplot as plt

f = open("data_file/line_fit_sample_data_with_delimiter_blank.txt", "r")

datas = []

lines = f.readlines()
for line in lines:
    datas.append(line[:-1].split(" "))

f.close()
```

```python
xs=[]
ys=[]

for d in datas:
    xs.append(d[0])
    ys.append(d[1])
```

```python
for i in range(len(datas)):
    plt.scatter(xs[i],ys[i])
plt.show()
```
